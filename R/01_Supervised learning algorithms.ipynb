# 01_Supervised learning algorithms

지도 학습은 훈련 데이터 세트라 불리는 사전 정의 된 데이터 세트에서 학습된 알고리즘을 의미  
훈련 데이터 세트는 일반적으로 입력 요소와 원하는 출력 요소 또는 신호로 구성된 두 요소 튜플임  
일반적으로 입력 요소는 벡터이며 지도 학습 알고리즘은 학습 데이터 세트를 사용하여 원하는 기능을 생성  
이렇게 생성 된 (또는 오히려 추정 된) 함수는 테스트 데이터로 더 잘 명명 된 새로운 데이터를 올바르게 맵핑하는 데 사용

---

잘 학습된 알고리즘은 보이지 않는 데이터의 출력을 합리적인 방법으로 정확하게 결정할 수 있음  
간단히 말하면, 일반화는 알고리즘이 올바른 방식으로 보이지 않는 데이터를 처리하기 위해 (제한된) 훈련 데이터에 기초하여 원하는 기능을 일반화하는 개념을 지칭함  
과적합(overfitting)은 정확히 일반화와 반대 개념임  
알고리즘은 함수를 추론하여 훈련 데이터 세트 (노이즈 포함)에 정확하게 매핑  
알고리즘에 의해 습득 된 함수가 새로운 보이지 않는 데이터에 대해 검사될 때 이것은 거대한 에러를 유발할 수 있음  

---

일반화와 과적합(overfitting)은 모두 입력 데이터의 무작위 오류 또는 노이즈를 중심으로 이루어지며 일반화는 노이즈의 영향을 최소화하려고 시도하지만, 과적합은 노이즈를 피팅함으로써 반대의 효과를 발휘  

---

머신러닝으로 해결해야 할 문제는 다음 단계로 나뉨  

1. 훈련 자료 준비 : 데이터 준비는 모든 기계 학습 알고리즘에서 가장 중요한 단계  
지도 학습은 분류된 입력 데이터 세트(주어진 입력에 해당하는 출력으로 구성된 데이터 세트)를 사용하므로이 단계가 더욱 중요  

2. 모델 준비 : 모델은 입력 데이터 집합 및 학습 된 패턴을 나타내며 모델 표현은 입력 기능 및 학습 알고리즘 자체와 같은 요소의 영향을 받음  
유추 된 함수의 정확성은이 표현이 어떻게 형성되는지에 따라 달라짐  

3. 알고리즘 선택 : 해결된 문제와 입력 정보를 기반으로 알고리즘을 선택하여 문제를 배우고 해결  

4. 검사 및 미세 조정 : 이것은 알고리즘이 입력 데이터 세트에서 실행되고 매개 변수가 원하는 수준의 출력을 얻도록 미세 조정되는 반복 단계임  
그런 다음 알고리즘은 성능을 평가하고 오류를 측정하기 위해 테스트 데이터 세트에서 테스트됨  

---

지도 학습 (supervised learning)하에 두 가지 주요 하위 범주는 다음과 같음

1. 회귀 기반의 기계 학습 : 학습 알고리즘은 우리가 몇 가지와 같은 정량적 인 질문에 대답하는 데 도움이 되는가? 또는 얼마나? 출력은 일반적으로 연속값이며 이 알고리즘은 훈련 데이터 및 형성된 모델을 기반으로 보이지 않는 새로운 데이터의 출력 값을 예측  
출력값은 이 경우 연속적(선형 회귀, 다 변수 회귀, 회귀 트리 등은 몇 가지 지도학습 회귀 알고리즘)

2. 분류 기반 기계 학습 : 객관적인 질문이나 예 또는 아니오 예측에 도움이되는 학습 알고리즘.  
예를 들어,이 구성 요소에 결함이 있는가? 또는이 종양이 암을 유발할 수 있는가? 보다 공식적으로이 알고리즘은 형성되는 학습 데이터 및 모델을 기반으로 보이지 않거나 새로운 데이터에 대한 클래스 레이블을 예측  
서포트 벡터 머신 (SVM), 의사 결정 트리, 랜덤 포레스트 등은 일반적으로 사용되는 지도학습 분류 알고리즘
