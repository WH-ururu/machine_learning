# 5. Predicting_Customer_Shopping_Trends_with_Market_Basket_Analysis

이 장에서는 오늘날 소매 업체, 상점 및 전자 상거래 장터가 직면 한 복잡한 문제 중 하나를 다룸  
현대적인 기술과 혁신의 출현으로 쇼핑은 웹이나 쇼핑 시설을 제공하는 전용 앱을 사용하여 실제 매장으로 모험하지 않고도 가정의 안락함에서 즐길 수 있는 상대적으로 즐겁고 즐거운 경험이되고 있음. 엄청난 수의 소매 업체, 상점, 마켓 플레이스 및 판매자가있어 경쟁이 치열하고 고객을 끌어 들이기 위해서는 소비자의 개인적 특성과 쇼핑 패턴에 대해 소비자가 모을 수있는 모든 데이터를 사용해야하고 기계 학습 기술을 사용하여 각 고객을 기반으로 가능한 한 개인화 된 쇼핑 서비스를 제공해야함 

각 사용자별로 쇼핑 경험을 개인화하는 데 기계 학습이 어떻게 도움이되는지 궁금 할 것입니다. 대답은 데이터와 알고리즘이라는 두 가지 측면에 있습니다. 이 두 가지를 조합하여 소매업자는 소비자가 가장 많이 팔리는 항목, 고객의 좋아하는 것과 싫어하는 것, 매출이 올라가고 내려갈 때가 가장 많은시기, 제품의 인기 급상승하는 조합을 파악할 수 있습니다 같은 제품에 대해 다른 소매 업체에서 제공하는 제품 리뷰 및 가격과 같은 다양한 정보를 제공합니다. 소매 업체는이 데이터를 집계하고 추세 제품을 분석하고 고객이 구매할 가능성이 높은 것을 예측하는 추천 엔진을 구축하고 관심 분야를 기반으로 고객에게 권장 사항을 제공하는 데 사용되는 다양한 기계 학습 알고리즘을 적용하는 자체 데이터 과학 팀을 운영합니다. 및 쇼핑 기록.

이 장에서 우리는 알고리즘이 고객이 구매할 가능성이 가장 높은 제품을 탐지하고 예측하기 위해 고객이 구매 한 제품 조합의 일반적인 패턴을 관찰하는 고객 쇼핑 거래 데이터에 중점을 두는 제품 기반 권장 사항에 초점을 맞출 것입니다 그리고 그들이 과거에 샀던 것. 이 장에서 다루게 될 주요 기술은 다음과 같습니다.

• 제품 비상 계획 매트릭스 평가

• 빈번 항목 집합 생성

• 연관 규칙 마이닝

연관 규칙과 패턴 마이닝을 사용한 추세 분석에는 고유 한 제한이 있습니다. 고객은 자신의 관심사, 구매 한 제품 및 평가 한 제품과 같은 속성을 기반으로 각 고객에 대해보다 개인화 된 쇼핑 경험을 제공하지 않습니다. 다음 장에서 권장 기반 엔진을 구축 할 때 제품 기반 및 사용자 기반 기능을 모두 고려한 사용자 기반 협업 필터링과 같은 알고리즘에 대해 살펴 보겠습니다.

가장 흥미로운 점은 Staples, Macy 's, Target, Amazon, Flipkart, Alibaba, 낙천, Lazada 및 기타 많은 소매 업체 및 전자 상거래 시장이 자체 데이터 과학 팀을 보유하고있어 다양한 우리가 앞서 논의한 문제를 포함하여 고객 쇼핑 거래, 제품 재고, 납품, SLA, 리뷰, 광고, 클릭률, 반송 률, 가격 데이터 및 기타 여러 출처에서 생성 된 모든 데이터를 사용합니다. 그들은이 데이터를 처리하고 기계 학습 알고리즘 기반 엔진에 데이터를 제공하여 데이터 기반 통찰력을 생성하여 비즈니스의 매출 및 이익을 늘립니다. 지금 이것은 분명히 현재 시장에서 인기있는 하나의 도메인입니다. 이제는 훌륭한 데이터 기반 의사 결정을 내리는 데 도움이되는 기계 학습 기술 및 알고리즘에 대해 살펴 보겠습니다.
